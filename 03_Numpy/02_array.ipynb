```python
# Different Operations on Arrays
```

```python
arr = np.array([[1,2,3],[4,5,6]])
print(arr,arr.shape,end="\n \n")

reshaped  = arr.reshape(3,2)
print(reshaped,reshaped.shape)
```

```python
flattened = arr.flatten() # 2D -> 1D
print(flattened,flattened.shape)
```

```python
# Indexing on 1D & 2D Array
```

```python
arr = np.array([1,2,3])
print(arr,arr.shape,end="\n \n")
print(f"arr[0] =  {arr[0]}")
```

```python
arr = np.array([[1,2,3],[4,5,6]])
print(arr,arr.shape,end="\n \n")
print(f"arr[0][1] =  {arr[0][1]}")
```

```python
#Fancy Indexing
arr = np.array([1,2,3,4,5,6,7])
idx=[0,5,2]
print(arr[idx])

# Boolean indexing
print(arr[arr>3])
print(arr[arr%2==0])
```

```python
#Slicing (start,end,step)
arr = np.array([1,2,3,4,5,6,7])
print(arr[1:4])
print(arr[1:])
print(arr[:4])
print(arr[::2])
```

```python
nums=[1,2,3,4,5,6]
sub_list=nums[1:3]
print(sub_list)
sub_list[0]=200
print(sub_list)
print(nums)
```

```python
# in numpy
arr=np.array([1,2,3,4,5])
sub_arr=arr[1:3]
print("sub_arr",sub_arr)
print("arr",arr)
sub_arr[0]=200
print("sub_arr",sub_arr)
print("arr",arr)
```

```python
#Copy in Numpy
arr=np.array([1,2,3,4,5])
sub_arr=arr[1:3].copy()
print("sub_arr",sub_arr)
print("arr",arr)
sub_arr[0]=200
print("sub_arr",sub_arr)
print("arr",arr)
```
