```python
#Mathematical Function
```

```python
#Aggregation
import numpy as np
```

```python
arr = np.array([1,2,3,4,5])
print(np.sum(arr))
print(np.prod(arr))
print(np.min(arr))
print(np.max(arr))
print(np.argmin(arr))#index val
print(np.argmax(arr)) #index val

print(np.mean(arr))
print(np.median(arr))
print(np.var(arr))
print(np.std(arr))
```

```python
#Power Function
```

```python
arr = np.array([1,2,3,4,5])
print(np.square(arr))
print(np.sqrt(arr))
print(np.pow(arr,3))
```

```python
#Logarithmic Function
```

```python
arr = np.array([1,2,3])
print(np.log(arr))
print(np.log2(arr))
print(np.log10(arr))
print(np.exp(arr))
```

```python
#Rounding
```

```python
print(np.round(3.14))
print(np.ceil(3.14))
print(np.floor(3.14))
print(np.trunc(3.14))
arr = np.array([1,2,3,4,1,2,3,4,5,6,6,54,433,2])
print(np.unique(arr))
print(np.sort(arr))
print(np.abs(arr))
```
