```python
import pandas as pd
```

```python
#Data Structure in Pandas
1. Series
2. DataFrame
```

```python
# 1. Series
```

```python
info = {
    "Name":["Rahul","faz","shradhha"],
    "CGPA":[7,8.5,9]
}
df = pd.DataFrame(info)
print(df)
```

```python
#Series
```

```python
s=pd.Series([1,2,3,4,5])
print(s)
print(type(s))
```

```python
#index
s=pd.Series([1,2,3,4,5])
print(s[0])
print(s[2])
```

```python
#label
s=pd.Series([11,22,33,42,57] , index = ["Adnan","Everest" ,"Bob","Charlie","Akik"])
print(s)
print(s["Adnan"])
print(s["Everest"])
print(s.index)
```

```python
#Vectorzed Operation
```

```python
s1=pd.Series([1,2,3,4,5])
s2=pd.Series([10,20,3,40,50])
print(s1+s2)
chnaged_s1=s1.drop(0)
print(s1)
print(chnaged_s1)
```

```python
#DataFrame
```

```python
# 1 .Dictionary Of List
info = {
    "Name" : ["Ada" ,"Book" ,"mf"] ,
    "Age": [23,24,25] ,
    "GPA" : [9.5, 8.8,9.2]
 }
df=pd.DataFrame(info)
print(df,type(df))
df
print("df.index and df.columns")
print(df.index)
print(df.columns)
```

```python
# 2 .List Of List
df = pd.DataFrame( [["Ada" , 23] ,["Book" , 25] ,["mf" , 19]] , columns=["Name","Age"])
print(df)
```

```python
# 3 .np Arrays 
```

```python
import numpy as np
np_arr = np.array([[1,2,3] , [4,5,6]])
df=pd.DataFrame(np_arr,columns=["a","b","c"])
print(df)
```
