```python
#Cleaning Data -> raw data.csv file 
```

```python
df=pd.read_csv("raw_data.csv")
```

```python
df.isnull()  # or  df.isna()
df.isna()
df.isna().sum()
df.dropna() # drop null values row 
df.fillna(0)
#df.dropna(axis=1) # drop null values col 
```

```python
cleaned_data=df.copy()
age_mean=cleaned_data["age"].mean()
cleaned_data["age"]=cleaned_data["age"].fillna(age_mean)
cleaned_data
```

```python
#Forward fill -> ffill()
df.ffill()
#Backward fill -> ffill()
df.bfill()
```

```python
# Handle Duplicate Data
df[["country","gender"]].duplicated()
df.drop_duplicates()
```
