```python
#Data Transformation
#apply , map , assign , replace(old,new)
```

```python
import pandas as pd
df=pd.read_csv("raw_data.csv")
```

```python
df2=df.copy()
df2["tax"]=df2["income"].apply(lambda x : "20%" if x >= 60000 else "10")
df2
```

```python
gender_map={"Male":"M","Female":"F","Unknown":"U"}
df2["gender"]=df2["gender"].map(gender_map)
df2
```

```python
df2=df2.assign(Salary = df2["income"] * 1.1)
df2
```

```python
df2["country"]=df2["country"].replace("USA","US")
df2
```

```python
#Rename
df2.columns=["ID","NAME","AGE","COUNTRY","GENDER","INCOME","TAX","SALARY"]
df2.columns = df2.columns.str.capitalize()
df2.rename(columns={"Salary" : "Salary "})
```

```python
df2.sort_values("Income")
sorted_df=df2.sort_values(["Income","Age"])
```

```python
df2
```

```python
df2=sorted_df.sort_index()
df2
```

```python
#reset
df2.reset_index()
df2=df2.reset_index(drop=True)
df2
```

```python
#rank()
sorted_df["Ranking"]=sorted_df["Income"].rank()
sorted_df
sorted_df["Ranking"]=sorted_df["Income"].rank(ascending=True,method='min')
sorted_df
```
