```python
import pandas as pd
df=pd.read_csv("raw_data.csv")
df
```

```python
df2=df.copy()
df2=df2.drop_duplicates()
df2["gender"]=df2["gender"].fillna("Female")
df2=df2.fillna(0)

df2=df2.sort_values("income")
df2=df2.reset_index(drop=True)
df2.to_csv("sorted_data.csv")
```

```python
#Groupby , mean/max/sum/min/count , agg()
```

```python
df2.groupby("country")["income"].mean()
```

```python
df2.groupby("country")["income"].min()
```

```python
df2.groupby("country")["income"].max()
```

```python
df2.groupby("gender")["income"].max()
```

```python
df2.groupby("gender")["income"].agg(["mean","max","min"])
```

```python
df2.groupby("gender")["income"].aggregate(["mean","max","min"])
```

```python
df2.groupby("gender")["income"].agg(avg_salary="mean",max_salary="max",min_salary="min")
```

```python
df.groupby("country").agg({
    "income":"max",
    "age" :"mean"
})
```

```python
df.groupby("country").agg(
    max_salary=("income","max"),
    avg_age=("age","mean")
)
```
