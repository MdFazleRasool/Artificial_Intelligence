```python
#Melt - Wide to Long Format
#parameter- melt(id_vars,value_vars,  var_name,   value_name)
#                keep    unpivot    new_col_name  new_col_value


# Pivot - long to wide format
#parameter- pivot(index,columns,values)
#                 keep
```

```python
df = pd.DataFrame({
    "country": ["USA", "USA", "India", "India"],
    "year": [2020, 2021, 2020, 2021],
    "sales": [100, 120, 90, 110],
    "profit": [20, 25, 18, 22],
    "tax" : [10,20,30,40]
})
```

```python
melted_df=df.melt(
    id_vars=["country","year"],
    value_vars=["sales","profit","tax"] ,
    var_name ="metrics",
    value_name="value"
)
melted_df
```

```python
pivot_df=melted_df.pivot(
    index = ["country","year"] ,
    columns = "metrics" ,
    values = "value"
)
pivot_df
```
