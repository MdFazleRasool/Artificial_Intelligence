```python
# Merging & Joining Data

import pandas as pd

df_customers = pd.DataFrame({
    "customer_id": [1, 2, 3, 4],
    "name": ["Adam", "Bob", "Charlie", "Dave"]
})

df_orders = pd.DataFrame({
    "order_id": [101, 102, 103, 104],
    "customer_id": [2, 1, 4, 5],
    "amount": [250, 120, 300, 180]
})
```

```python
df_customers
```

```python
df_orders
```

```python
pd.merge(df_customers,df_orders,on="customer_id")
```

```python
pd.merge(df_customers,df_orders,on="customer_id" , how="left")
```

```python
pd.merge(df_customers,df_orders,on="customer_id" , how="right")
```

```python
pd.merge(df_customers,df_orders,on="customer_id" , how="outer")
```

```python
# Data Concetation :-  Stacking DataFrames on top of each other (rows) or side  by side (columns) 
```

```python
df1 = pd.DataFrame({
    "id": [1, 2, 3],
    "name": ["Adam", "Eve", "Bob"]
})

df2 = pd.DataFrame({
    "id": [4, 5, 6],
    "name": ["Charlie", "Fester", "Gilian"]
})
```

```python
df1
```

```python
df2
```

```python
pd.concat([df1,df2]) # row
```

```python
pd.concat([df1,df2] , axis = 1) # col
```

```python
pd.concat([df1,df2] , axis = 1 ,ignore_index=True) # col
```

```python
pd.concat([df1,df2] , ignore_index=True) #row
```
