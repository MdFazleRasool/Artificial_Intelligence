```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://www.scrapethissite.com/pages/simple/"
res = requests.get(url)
```

```python
print(res.status_code)
```

```python
if res.status_code == 200:
    print(res.text)
    print(res.content)
    print(res.headers)
    
```

```python
with open("scraped_data/data1.html", "w", encoding="utf-8") as f:
    f.write(res.text)
import os
print(os.path.getsize("scraped_data/data1.html"))
```

```python
# Beautiful Soup
with open ("scraped_data/data1.html", "r") as f :
    html_content = f.read()

soup = BeautifulSoup(html_content,"lxml")
```

```python
print(len(html_content))
print(soup)
print(len(str(soup)))
```

```python
# Methods
soup.find('h3')
```

```python
all_h3=soup.find_all('h3')
type(all_h3)
```

```python
for h3 in all_h3 :
    print(h3.get_text(strip = True) , end =" :- ")
    #  print(h3.find_parent("div")["class"])
    print(h3.find_next("div").select("span.country-population")[0].get_text(strip = True))
```

```python
print(h3.find_next("div"))
```

```python
print(h3.find_next("div").select("span.country-population")[0].get_text())
```

```python

all_countries = []
for h3 in all_h3 :
    name = h3.get_text(strip = True) 
    population = h3.find_next("div").select("span.country-population")[0].get_text(strip = True)
    all_countries.append([name,population])
```

```python
all_countries
```

```python
import pandas as pd
df = pd.DataFrame(all_countries,columns=["name","population"])
print(df)
```

```python
df.to_json("cleaned_data/data.json",index=False)
df.to_csv("cleaned_data/data.csv",index=False)
```
