```python
# Corrected Best Picture Data (2008 - 2012)
oscar_movies = [
    "The Dark Knight",
    "The Hurt Locker",
    "The King's Speech",
    "The Artist",
    "Argo"
]
years = [2008, 2009, 2010, 2011, 2012]
oscar_revenue = [1050, 170, 427, 133, 232] # in $M

# plot
plt.title("Oscar Movbies revenue in each year")
plt.xlabel("years")
plt.ylabel("revenue (in $M)")
```

```python
  # Non-Oscar Winner Movies
non_oscar_movies = ["Slumdog Millionaire","Avatar","Inception","Hugo","Lincoln"]
non_oscar_revenue = [378, 2788, 829, 185, 275] # in $M
```

```python
plt.plot(years,oscar_revenue,label="oscar Movies")
plt.plot(years,non_oscar_revenue,label="Non oscar Movies")
plt.legend(loc="right")
plt.title("Oscar Movies revenue V/s Non-Oscar Movies revenue ")
plt.xlabel("years")
plt.ylabel("revenue (in $M)")
```

```python
# Format Strings - ( marker , line , color)
```

```python
plt.plot(years,oscar_revenue,"o--r",label="oscar Movies")
plt.plot(years,non_oscar_revenue,"<-y ",label="Non oscar Movies")
plt.legend(loc="right")
plt.title("Oscar Movies revenue V/s Non-Oscar Movies revenue ")
plt.xlabel("years")
plt.ylabel("revenue (in $M)")
```

```python
plt.plot(years,oscar_revenue,color="#000",label="oscar Movies")
plt.plot(years,non_oscar_revenue,marker="o",linewidth="5",label="Non oscar Movies")
plt.legend(loc="right")
plt.title("Oscar Movies revenue V/s Non-Oscar Movies revenue ")
plt.xlabel("years")
plt.ylabel("revenue (in $M)")
```
