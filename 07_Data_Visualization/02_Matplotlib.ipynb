```python
import matplotlib.pyplot as plt

movies = [
    "The Dark Knight",
    "The Hurt Locker",
    "The King's Speech",
    "The Artist",
    "Argo"
]

years = [2008, 2009, 2010, 2011, 2012]
oscar_revenue = [1005, 170, 427, 133, 232]  # in $M

plt.bar(years, oscar_revenue)

plt.title("Revenue in each Year for Oscar Movies" )
plt.xlabel("Year")
plt.ylabel("Revenue (in $M)")


for i in range(len(years)):
    plt.text(years[i],oscar_revenue[i]+10,str(oscar_revenue[i]),ha="center")
plt.ylim(0,max(oscar_revenue)+200)
plt.xlim(min(years)-1,max(years)+1)
plt.show()
```

```python
  # Non-Oscar Winner Movies
non_oscar_movies = [
    "Slumdog Millionaire",
    "Avatar",
    "Inception",
    "Hugo",
    "Lincoln"
]
non_oscar_revenue = [378, 2788, 829, 185, 275] # in $M

oscar_movies = [
    "The Dark Knight",
    "The Hurt Locker",
    "The King's Speech",
    "The Artist",
    "Argo"
]
oscar_revenue = [1005, 170, 427, 133, 232]  # in $M

years = [2008, 2009, 2010, 2011, 2012]

width = 0.4
plt.bar(years,oscar_revenue)
plt.bar(years,non_oscar_revenue)

plt.title("Oscar Movies V/s Non-Oscar Movies")
plt.xlabel("years")
plt.ylabel("Revenue ($M)")
```

```python
import numpy as np
width = 0.4
years=[2008,2009,2010,2011,2012]    
x=np.arange(len(years))    
plt.bar(x-width/2,oscar_revenue,width , label="Oscar_Movies")
plt.bar(x+width/2,non_oscar_revenue,width,  label="Non_Oscar_Movies")
plt.legend()
plt.title("Oscar Movies V/s Non-Oscar Movies")
plt.xlabel("years")
plt.ylabel("Revenue ($M)")
plt.xticks(x,years)
plt.show()
```

```python
# Horizontal Bar Charts
non_oscar_movies = [
    "Slumdog Millionaire",
    "Avatar",
    "Inception",
    "Hugo",
    "Lincoln"
]
non_oscar_revenue = [378, 2788, 829, 185, 275] # in $M
plt.barh(non_oscar_movies,non_oscar_revenue)
plt.xlabel("Revenue")
plt.ylabel("non_oscar_movies")
```
