```python
import matplotlib.pyplot as plt
```

```python
# Scatter plot
people = ["Person A", "Person B", "Person C", "Person D", "Person E",
          "Person F", "Person G", "Person H", "Person I", "Person J"]
age = [22, 25, 30, 35, 40, 45, 50, 55, 60, 65]
bp = [110, 115, 120, 122, 125, 130, 135, 123, 145, 150]

colors = ["red", "purple", "yellow", "blue", "pink", "red", "purple", "yellow", "blue", "pink"]

plt.scatter(age,bp,color=colors)
plt.title("Age vs Bp")
plt.xlabel("Age")
plt.ylabel("Blood Pressure")
#plt.grid()
```

```python
# Customization in scatter plots 
# 1. Color(color) , 2. Size(s) , 3. Marker(marker) , 4. Transparency(alpha) , 5. Colormap(cmap) , 6. Scale (colorbar) 

colors = ["green" if x < 135 else "red" for x in bp]
plt.scatter(age,bp,color=colors , s = bp , marker="^")
plt.title("Age vs Bp")
plt.xlabel("Age")
plt.ylabel("Blood Pressure")
```

```python

colors = ["green" if x < 135 else "red" for x in bp]
plt.scatter(age,bp, alpha = 0.7 , cmap="OrRd",c=bp , s=bp)
plt.title("Age vs Bp")
plt.xlabel("Age")
plt.ylabel("Blood Pressure")
plt.colorbar(label="Bp")
plt.grid()
plt.show()
```

```python
# Annotation
colors = ["green" if x < 135 else "red" for x in bp]
plt.scatter(age,bp, alpha = 0.7 , cmap="OrRd",c=bp , s=bp)
plt.title("Age vs Bp")
plt.xlabel("Age")
plt.ylabel("Blood Pressure")
plt.colorbar(label="Bp")
plt.grid()
plt.ylim(min(bp)-5,max(bp)+5)
plt.xlim(min(age)-5,max(age)+5)

for i in range(len(people)):
    plt.annotate(people[i] , xy=(age[i],bp[i]) , xytext = (age[i],bp[i]))
plt.show()
```
