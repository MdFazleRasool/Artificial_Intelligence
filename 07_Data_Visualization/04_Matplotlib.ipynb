```python
import matplotlib.pyplot as plt
```

```python
# Pie Charts
expenses = ["Salaries", "Rent", "Marketing", "R&D", "Miscellaneous"]
amounts = [500, 230, 180, 130, 80]

plt.style.use("ggplot")
plt.pie(amounts, labels=expenses)
plt.show()
```

```python
# Custamiztion in Pie Chart
plt.style.use("ggplot")
plt.pie(amounts, labels=expenses,autopct="%1.1f%%")
plt.show()  
```

```python
plt.style.use("ggplot")
#plt.pie(amounts, labels=expenses,autopct="%1.1f%%")
# Generating the pie chart with customized slice properties
plt.pie(
    amounts, 
    labels=expenses, 
    autopct="%1.1f%%",
    wedgeprops={
        "edgecolor": "black",
        "linewidth": 1,
        "linestyle": "--",
        "fill": True
    }
)
plt.show()
```

```python
plt.style.use("ggplot")
# Generating the pie chart with customized slice properties
explode=[0,0.2,0,0,0]
plt.pie(
    amounts, 
    labels=expenses, 
    autopct="%1.1f%%",
    wedgeprops={
        "edgecolor": "black",
        "linewidth": 1,
        "linestyle": "--",
        "fill": True
    },
    explode=explode,
    startangle=90,
    shadow=True
)
plt.show()
```
