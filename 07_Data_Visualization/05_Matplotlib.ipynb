```python
# Histoghram - Preferred for numerical data
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(0)
scores = np.random.normal(70,10,100)
plt.hist(scores,color="skyblue",edgecolor="black",bins=15)
plt.show()
```

```python
# Multiple datasets

# multiple datasets
legit_transactions = [
    2.99, 5.49, 8.99, 12.50, 14.99, 19.99, 23.45, 29.99, 34.99, 39.50,
    45.00, 49.99, 55.25, 60.00, 75.99, 89.99, 120.50, 150.00, 199.99,
    249.99, 300.75, 450.00, 600.00, 850.00, 1200.00
]

fraud_transactions = [
    50, 100, 150, 200, 300, 500, 500, 750, 1000, 1000,
    1200, 1500, 1500, 2000, 2500, 3000, 3000
]

# Transaction Amount Distrbution - Legit vs Fraud
plt.hist(legit_transactions,bins=10,label="legit_transactions",edgecolor="black",alpha=0.5)
plt.hist(fraud_transactions,bins=10,label="fraud_transactions",edgecolor="black",alpha=0.5)
plt.title("Transaction Amount Distrbution - Legit vs Fraud")
plt.xlabel("Amount (in $)")
plt.ylabel("Frequency")
plt.legend()
plt.grid()
plt.show()
```

```python
#axvline

marks = [
    25, 38, 42, 55, 61, 67, 72, 78, 83, 88,
    91, 95, 47, 53, 59, 64, 69, 74, 79, 85,
    92, 97, 33, 45, 58, 62, 71, 76, 81, 89
]

plt.hist(marks, color="skyblue", edgecolor="black")
plt.ylabel("Number of students")
plt.xlabel("Marks")
plt.axvline(
    x=33,
    label="Passing marks (33)",
    color="red",
    linewidth=2,
    linestyle="--"
)
plt.show()
```
