```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = [7,8,5,6,9,4,10,12,15]
plt.boxplot(data)
plt.show()
```

```python
# multiple datasets
group1 = np.random.normal(50, 10, 100)
group2 = np.random.normal(60, 15, 100)

plt.boxplot([group1, group2], tick_labels=["group1", "group2"])
plt.grid()
plt.show()
```

```python
# Horizontal box plot
group1 = np.random.normal(50, 10, 100)
group2 = np.random.normal(60, 15, 100)

plt.boxplot(
    [group1, group2], 
    tick_labels=["group1", "group2"], 
    vert = False,
    showmeans=True ,
    whis= 2.0
)
plt.grid()
plt.show()
```
