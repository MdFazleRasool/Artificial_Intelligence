```python
# Relation plots -> (scatter , line)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset("tips")
print(type(tips))
print(tips.head())
```

```python
sns.scatterplot(
     data=tips ,
     x = "total_bill" ,
     y = "tip" ,
     hue  = "time"
)
```

```python
# line plot
flights = sns.load_dataset("flights")
# print(flights)
sns.lineplot(
     data=flights ,
     x = "year" ,
     y = "passengers" 
)
```

```python
sns.lineplot(
     data=flights ,
     x = "year" ,
     y = "passengers" ,
        hue = "month"
)
```
