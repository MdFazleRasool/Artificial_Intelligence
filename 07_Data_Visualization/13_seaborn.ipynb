```python
# Categorical plots --> (Bar plots , Box plots)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips=sns.load_dataset("tips")
print(tips.head())
```

```python
# Bar Plot
sns.barplot(
    data=tips,
    x="day",
    y="tip",
    hue="sex"
)
```

```python
# Box plot
sns.boxplot(
    data=tips,
    x="day",
    y="tip",
    hue="sex"
)
```
