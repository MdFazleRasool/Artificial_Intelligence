```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips= sns.load_dataset("tips")
tips.head()
```

```python
fig,ax = plt.subplots()
sns.lineplot(
    data = tips ,
    x="day",
    y="total_bill",
    hue="sex",
    marker="o",
    errorbar=None ,
    ax=ax
)
ax.set_title("Avg bill value for each day")
ax.set_xlabel("Days")
ax.set_ylabel("Total Bill Amount")
fig.tight_layout()
```
