```python
# Function :- Each input has only one output. 
# Example :- f(x) = 2x + 3, where x is the input    
# f(1) = 2*1 + 3 = 5, f(2) = 2*2 + 3 = 7, f(3) = 2*3 + 3 = 9
# In this example, the function f takes an input x and produces an output by applying the formula 2x + 3. Each value of x will yield a unique output, demonstrating the concept of a function.  
# Range :- The set of all possible output values of a function.
# Domain :- The set of all possible input values of a function.                                                                
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define x values (x >= 0 for sqrt)
x = np.linspace(0, 100, 10000)
y=np.sqrt(x)
# Plot the function
plt.plot(x, y, label='y = sqrt(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Function: y = sqrt(x)')
plt.legend()
plt.grid(True)
plt.show()
```
