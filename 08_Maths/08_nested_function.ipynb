```python
# Nested Function or Composite Function :- A function that is defined inside another function.
# Example :-f(g(x)) = 2(g(x)) + 3, where g(x) = x^2
# f(g(x)) = 2(x^2) + 3 = 2x^2 + 3
# In this example, g(x) is a function that takes an input x and produces an output by applying the formula x^2.
#                                   
```

```python
# Graph of sinx
```

```python
import numpy as np
import matplotlib.pyplot as plt
# Create x values from -2π to 2π
x = np.linspace(-2 * np.pi, 2 * np.pi, 100)
y = np.sin(x)

# Plot the graph
plt.plot(x, y, label='y = sin(x)', color='blue')

# Highlight x=0 and y=0 lines
plt.axvline(0, color='black', linewidth=1.5, linestyle='--', label='x = 0')
plt.axhline(0, color='black', linewidth=1.5, linestyle='--', label='y = 0')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Graph of y = sin(x)')
plt.grid(True)
plt.legend()

plt.show()
```
