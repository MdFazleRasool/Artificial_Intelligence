```python
#Machine learning :- ML is a part of computer science where we teach computers to learn from data, find patterns & make decision or predictions.
```

```python
# Types of machine learning:
#1. Supervised learning . 
#2. Unsupervised learning .
#3. Reinforcement learning .
```

```python
# Supervised Learning:- Supervised ML trains algorithms on labeled datasets, meaning each input has a known correct output, to learn patterns and predict outcomes for new data.
```

```python
# Supervised Learning
# Types of problems
# 1 .Regression - output is a continuous numerical value
    #e.g:- house price prediction , stock price prediction .
# 2 .Classification - output is a category or class
 # eg:- spam detection , fraud detection .
```

```python
# Types of classification problems
# 1. Binary classification - output is one of two classes
# 2. Multiclass classification - output is one of multiple classes 
# 3. Multilabel classification - output can belong to multiple classes simultaneously
```

```python
# scikit-learn :-  is an open-source ML library that provides simple and efficient tools for data mining and data analysis. It is built on top of NumPy, SciPy, and matplotlib.
# It also supports  supervised and unsupervised learning .
# Algorithms:- 1. Linear regression -> regression problems
#            2. Logistic regression -> classification problems
#            3. K-nearest neighbors (KNN) ->both
#            4. Decision trees
#            5. Naive Bayes           -> classification problems
#            6. Support vector machines (SVM)
```
